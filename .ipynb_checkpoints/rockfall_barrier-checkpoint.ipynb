# Rockfall Barrier

```python
from math import pi, cos, sin, tan

import numpy as np
from matplotlib import pyplot as plt
from matplotlib.patches import Polygon
from matplotlib.text import Text
from matplotlib.text import Annotation

from intersect import intersection
```

```python
# input data
land_slope = 24.1 # pente du terrain naturel en degré - strictement supérieure à 0
upstream_slope = 32 # pente amont en degré - strictement supérieure à land_slope
pit_width = 5 # largeur de fosse en mètre
useful_height = 4 # hauteur utile en mètre - non nulle
siding_slope = 70 # pente du parement en degré
siding_width = 1 # largeur du parement en crête
siding_sole = 0.5 # hauteur de la semelle du parement
crest_width = 3.5 # largeur de crête en mètre
downstream_slope = 32 # pente aval en degré - strictement supérieure à land_slope
height_above_land = 3.6 # hauteur de l'axe de la crête au-dessus du TN - non nulle
```

```python
# conversion
land_slope = abs(land_slope) * pi / 180
upstream_slope = abs(upstream_slope) * pi / 180
pit_width = abs(pit_width)
useful_height = abs(useful_height)
siding_slope = abs(siding_slope) * pi / 180
siding_width = abs(siding_width)
siding_sole = abs(siding_sole)
crest_width = abs(crest_width)
downstream_slope = abs(downstream_slope) * pi / 180
height_above_land = abs(height_above_land)
```

```python
# check input values
valid = True

if abs(land_slope) == 0:
    valid = False
elif upstream_slope < land_slope and downstream_slope < land_slope:
    valid = False
elif useful_height == 0 or height_above_land == 0:
    valud = False

valid
```

```python
h = (0, height_above_land)
i = (h[0] + crest_width/2, h[1])
d = (i[0] - crest_width, i[1])
e = (d[0] + siding_width, d[1])
f = (e[0], e[1] - useful_height - siding_sole)
c = (d[0] - useful_height / tan(siding_slope), d[1] - useful_height)
g = (c[0], c[1] - siding_sole)
b = (c[0] - pit_width, c[1])
```

```python
# check if 'c' and 'd' are on either side of the natural land
F = lambda x: -tan(land_slope) * x

valid = True

if d[1] < F(d[0]) or c[1] > F(c[0]):
    valid = False
    
valid
```

```python
# upstream intersection
A = np.array([[tan(land_slope), 1], [tan(upstream_slope), 1]])
B = np.array([0, b[1] + tan(upstream_slope) * b[0]])
X = np.linalg.solve(A, B)

a = (X[0], X[1])
```

```python
# downstream intersection
A = np.array([[tan(land_slope), 1], [tan(downstream_slope), 1]])
B = np.array([0, i[1] + tan(downstream_slope) * i[0]])
X = np.linalg.solve(A, B)

j = (X[0], X[1])
```

```python
# upstream siding intersection
xy_cd = [c, d]
x_cd = [xy[0] for xy in xy_cd]
y_cd = [xy[1] for xy in xy_cd]

xy_aj = [a, j]
x_aj = [xy[0] for xy in xy_aj]
y_aj = [xy[1] for xy in xy_aj]

x_l, y_l = intersection(x_cd, y_cd, x_aj, y_aj)
l = (float(x_l), float(y_l))
```

```python
# downstream siding intersection
xy_gf = [g, f]
x_gf = [xy[0] for xy in xy_gf]
y_gf = [xy[1] for xy in xy_gf]

x_k, y_k = intersection(x_gf, y_gf, x_aj, y_aj)

k_in_gf = True

if len(x_k) == 0:
    k_in_gf = False
    xy_fe = [f, e]
    x_fe = [xy[0] for xy in xy_fe]
    y_fe = [xy[1] for xy in xy_fe]
    
    x_k, y_k = intersection(x_fe, y_fe, x_aj, y_aj)
    
    k = (float(x_k), float(y_k))
else:
    k = (float(x_k), float(y_k))
```

```python
# function to calculate area of a polygon
def polygon_area(xy):
    area = 0
    
    x = [xy[0] for xy in xy]
    y = [xy[1] for xy in xy]
    
    for i in range(len(xy)):
        if i < len(xy) - 1:
            area += (x[i] * y[i+1] - x[i+1] * y[i])
        else:
            area += (x[-1] * y[0] - x[0] * y[-1])
            
    return abs(area) / 2
```

```python
# area
excavation_area = polygon_area([a, b, c, l])

if k_in_gf:
    backfill_area = polygon_area([k, f, e, i, j])
    excavation_siding_area = polygon_area([l, c, g, k])
    backfill_siding_area = polygon_area([l, k, f, e, d])
else:
    backfill_area = polygon_area([k, e, i, j])
    excavation_siding_area = polygon_area([l, c, g, f, k])
    backfill_siding_area = polygon_area([l, k, e, d])
    
siding_area = excavation_siding_area + backfill_siding_area
```

```python
# global width
width = j[0] - a[0]
```

```python
# results
print(f"Global width = {round(width, 2)} m")
print("---")
print(f"Siding area = {round(excavation_siding_area, 2)} (excavation) + {round(backfill_siding_area ,2)} (backfill) = {round(excavation_siding_area + backfill_siding_area ,2)} m^2")
print("---")
print(f"Excavation = {round(excavation_area ,2)} m^2")
print("---")
print(f"Backfill = {round(backfill_area ,2)} m^2")
```

```python
# plot
fig, ax = plt.subplots(figsize=(29.7/2.54,21/2.54))

ax.axis('equal')
ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)
ax.get_xaxis().set_ticks([])
ax.get_yaxis().set_ticks([])

# natural land
xy_land = [a, j]
x_land = [xy[0] for xy in xy_land]
y_land = [xy[1] for xy in xy_land]

land, = ax.plot(x_aj, y_aj)
land.set_label(f"Terrain Naturel ({round(land_slope * 180 / pi, 1)}°)")
land.set_linestyle('-')
land.set_linewidth(1)
land.set_color("Green")
land.set_alpha(1)
land.set_zorder(3)
land.set_visible(True)

# rockfall barrier
xy_barrier = [a, b, c, d, i, j]
x_barrier = [xy[0] for xy in xy_barrier]
y_barrier = [xy[1] for xy in xy_barrier]

barrier, = ax.plot(x_barrier, y_barrier)
barrier.set_label("Projet")
barrier.set_color("Red")
barrier.set_linestyle('-')
barrier.set_linewidth(1)
barrier.set_color("Red")
barrier.set_alpha(1)
barrier.set_zorder(3)
barrier.set_visible(True)

# siding
xy_siding = [c, d, e, f, g, c]
x_siding = [xy[0] for xy in xy_siding]
y_siding = [xy[1] for xy in xy_siding]

xy_siding = np.array([x_siding, y_siding]).T

siding = Polygon(xy_siding)
siding.set_label(f"Parement ({round(siding_area, 2)} $m^2$)")
siding.set_linewidth(0)
#siding.set_color("Red")
siding.set_color("Black")
siding.set_alpha(0.6)
siding.set_zorder(2)
siding.set_visible(True)

ax.add_patch(siding)

# excavation
xy_excavation = [a, b, c, l]
x_excavation = [xy[0] for xy in xy_excavation]
y_excavation = [xy[1] for xy in xy_excavation]

xy_excavation = np.array([x_excavation, y_excavation]).T

excavation = Polygon(xy_excavation)
excavation.set_label(f"Déblai ({round(excavation_area + excavation_siding_area, 2)} $m^2$)")
excavation.set_linewidth(0)
excavation.set_color("Yellow")
excavation.set_alpha(0.6)
excavation.set_zorder(1)
excavation.set_visible(True)

ax.add_patch(excavation)

# backfill
if k_in_gf:
    xy_backfill = [k, f, e, i, j, k]
else:
    xy_backfill = [k, e, i, j, k]

x_backfill = [xy[0] for xy in xy_backfill]
y_backfill = [xy[1] for xy in xy_backfill]

xy_backfill = np.array([x_backfill, y_backfill]).T

backfill = Polygon(xy_backfill)
#backfill.set_label(f"Remblai technique ({round(backfill_area + backfill_siding_area, 2)} $m^2$)")
backfill.set_label(f"Remblai ({round(backfill_area, 2)} $m^2$)")
backfill.set_linewidth(0)
backfill.set_color("Red")
backfill.set_alpha(0.6)
backfill.set_zorder(1)
backfill.set_visible(True)

ax.add_patch(backfill)

# annotations
# slope_upstream
slope_upstream = Text(0,
                      0,
                      "",
                      rotation=0,
                      rotation_mode="anchor",
                      horizontalalignment="center",
                      verticalalignment="top")

slope_upstream.set_text(f"\nTalus amont ({round(upstream_slope * 180 / pi, 1)}°)")
slope_upstream.set_x((a[0] + b[0]) / 2)
slope_upstream.set_y((a[1] + b[1]) / 2)
slope_upstream.set_fontsize(10)
slope_upstream.set_color("Black")
slope_upstream.set_rotation(-upstream_slope * 180 / pi)
slope_upstream.set_alpha(1)
slope_upstream.set_zorder(1)
slope_upstream.set_visible(True)

ax.add_artist(slope_upstream)

# slope_downstream
slope_downstream = Text(0,
                        0,
                        "",
                        rotation=0,
                        rotation_mode="anchor",
                        horizontalalignment="center",
                        verticalalignment="bottom")

slope_downstream.set_text(f"Talus aval ({round(downstream_slope * 180 / pi, 1)}°)\n")
slope_downstream.set_x((i[0] + j[0]) / 2)
slope_downstream.set_y((i[1] + j[1]) / 2)
slope_downstream.set_fontsize(10)
slope_downstream.set_color("Black")
slope_downstream.set_rotation(-downstream_slope * 180 / pi)
slope_downstream.set_alpha(1)
slope_downstream.set_zorder(1)
slope_downstream.set_visible(True)

ax.add_artist(slope_downstream)

#pit
pit = Text(0,
           0,
           "",
           rotation=0,
           rotation_mode="anchor",
           horizontalalignment="center",
           verticalalignment="top")

pit.set_text(f"\nFosse ({round(pit_width, 2)} m)\n")
pit.set_x((b[0] + c[0]) / 2)
pit.set_y((b[1] + c[1]) / 2)
pit.set_fontsize(10)
pit.set_color("Black")
pit.set_alpha(1)
pit.set_zorder(1)
pit.set_visible(True)

ax.add_artist(pit)

# crest
crest = Annotation("",
                   xy=(0, 0),
                   xytext=(0, 0),
                   xycoords='data',
                   rotation=0,
                   horizontalalignment='center',
                   verticalalignment='bottom',
                   arrowprops=dict(arrowstyle='->'))

crest.arrow_patch.set_arrowstyle("]->")
crest.arrow_patch.set_linestyle("-")
crest.arrow_patch.set_linewidth(1)
crest.arrow_patch.set_color("Black")
crest.arrow_patch.set_alpha(1)
crest.arrow_patch.set_zorder(1)

crest.set_text(f"Crête ({round(crest_width, 2)} m)")
crest.xy = (h[0], h[1])
crest.set_x(h[0])
crest.set_y(h[1] + 0.75*crest_width)
crest.set_color("Black")
crest.set_fontsize(10)
crest.set_alpha(1)
crest.set_zorder(1)
crest.set_visible(True)

ax.add_artist(crest)

# siding
siding = Annotation("",
                    xy=(0, 0),
                    xytext=(0, 0),
                    xycoords='data',
                    rotation=0,
                    horizontalalignment='center',
                    verticalalignment='top',
                    arrowprops=dict(arrowstyle='->'))

siding.arrow_patch.set_arrowstyle("]->")
siding.arrow_patch.set_linestyle("-")
siding.arrow_patch.set_linewidth(1)
siding.arrow_patch.set_color("Black")
siding.arrow_patch.set_alpha(1)
siding.arrow_patch.set_zorder(1)

siding.set_text(f"Parement ($h_u$ = {round(useful_height, 2)} m | {round(siding_slope * 180 / pi, 1)}°)")
siding.xy = ((f[0] + g[0]) / 2, g[1])
siding.set_x((f[0] + g[0]) / 2)
siding.set_y(g[1] - 0.75*useful_height)
siding.set_color("Black")
siding.set_fontsize(10)
siding.set_alpha(1)
siding.set_zorder(1)
siding.set_visible(True)

ax.add_artist(siding)

# upstream_width
upstream_width = Text(0,
                     0,
                     "",
                     horizontalalignment='center',
                     verticalalignment='top')

upstream_width_arrow = Annotation("",
                                  xy=(0, 0),
                                  xytext=(0, 0),
                                  xycoords='data',
                                  arrowprops=dict(arrowstyle='->'))

upstream_width_arrow.arrow_patch.set_arrowstyle('<->')
upstream_width_arrow.arrow_patch.set_linestyle('-')
upstream_width_arrow.arrow_patch.set_linewidth(1)
upstream_width_arrow.arrow_patch.set_color("Gray")
upstream_width_arrow.arrow_patch.set_alpha(1)
upstream_width_arrow.arrow_patch.set_zorder(2)        
upstream_width_arrow.xy = (a[0], b[1])
upstream_width_arrow.set_x(b[0])
upstream_width_arrow.set_y(b[1])
upstream_width_arrow.set_alpha(1)
upstream_width_arrow.set_zorder(1)
upstream_width_arrow.set_visible(True)

upstream_width.set_text(f"\n{round(b[0] - a[0], 2)} m")
upstream_width.set_x((a[0] + b[0]) / 2)
upstream_width.set_y(b[1])
upstream_width.set_fontsize(10)
upstream_width.set_color("Gray")
upstream_width.set_alpha(1)
upstream_width.set_zorder(1)
upstream_width.set_visible(True)

ax.add_artist(upstream_width_arrow)
ax.add_artist(upstream_width)

# downstream_slope
downstream_width = Text(0,
                        0,
                        "",
                        horizontalalignment='center',
                        verticalalignment='bottom')

downstream_width_arrow = Annotation("",
                                    xy=(0, 0),
                                    xytext=(0, 0),
                                    xycoords='data',
                                    arrowprops=dict(arrowstyle='->'))

downstream_width_arrow.arrow_patch.set_arrowstyle('<->')
downstream_width_arrow.arrow_patch.set_linestyle('-')
downstream_width_arrow.arrow_patch.set_linewidth(1)
downstream_width_arrow.arrow_patch.set_color("Gray")
downstream_width_arrow.arrow_patch.set_alpha(1)
downstream_width_arrow.arrow_patch.set_zorder(2)        
downstream_width_arrow.xy = (i[0], i[1])
downstream_width_arrow.set_x(j[0])
downstream_width_arrow.set_y(i[1])
downstream_width_arrow.set_alpha(1)
downstream_width_arrow.set_zorder(1)
downstream_width_arrow.set_visible(True)

downstream_width.set_text(f"{round(j[0] - i[0], 2)} m\n")
downstream_width.set_x((i[0] + j[0]) / 2)
downstream_width.set_y(i[1])
downstream_width.set_fontsize(10)
downstream_width.set_color("Gray")
downstream_width.set_alpha(1)
downstream_width.set_zorder(1)
downstream_width.set_visible(True)

ax.add_artist(downstream_width_arrow)
ax.add_artist(downstream_width)

# global_width
global_width = Text(0,
                    0,
                    "",
                    horizontalalignment='center',
                    verticalalignment='top')

global_width_arrow = Annotation("",
                                xy=(0, 0),
                                xytext=(0, 0),
                                xycoords='data',
                                arrowprops=dict(arrowstyle='->'))

global_width_arrow.arrow_patch.set_arrowstyle('<->')
global_width_arrow.arrow_patch.set_linestyle('-')
global_width_arrow.arrow_patch.set_linewidth(1)
global_width_arrow.arrow_patch.set_color("Gray")
global_width_arrow.arrow_patch.set_alpha(1)
global_width_arrow.arrow_patch.set_zorder(2)        
global_width_arrow.xy = (a[0], j[1])
global_width_arrow.set_x(j[0])
global_width_arrow.set_y(j[1])
global_width_arrow.set_alpha(1)
global_width_arrow.set_zorder(1)
global_width_arrow.set_visible(True)

global_width.set_text(f"\n{round(width, 2)} m")
global_width.set_x((a[0] + j[0]) / 2)
global_width.set_y(j[1])
global_width.set_fontsize(10)
global_width.set_color("Gray")
global_width.set_alpha(1)
global_width.set_zorder(1)
global_width.set_visible(True)

ax.add_artist(global_width_arrow)
ax.add_artist(global_width)

# © onf-rtm
onf_rtm = Text(0.995,
               0.005,
               "",
               rotation=0,
               rotation_mode="anchor",
               horizontalalignment="right",
               verticalalignment="bottom",
               transform=ax.transAxes)

onf_rtm.set_text("© Rockfall-Barrier | ONF RTM")
onf_rtm.set_fontsize(8)
onf_rtm.set_color("Gray")

ax.add_artist(onf_rtm)

# legend
ax.legend(loc='upper center',
          ncol=5,
          fontsize=10,
          frameon=True,
          bbox_to_anchor=(0.5, 1),
          bbox_transform=ax.transAxes)

fig.tight_layout()
plt.savefig("./png/merlon_P1.png")
plt.show()
```
